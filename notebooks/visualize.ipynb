```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
import numpy as np
```

```python
df_cleand = pd.read_csv(r"../data/cleaned_data.csv")
df_selected = pd.read_csv(r"../data/features_selected.csv")

```

```python
# Generate Correlation Heatmap
plt.figure(figsize=(10, 6))
corr_matrix = df_cleand.corr()
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
# 1. Statistical Summary
display(df_selected.describe())
```

```python
# 2. Histograms for numerical features
df_selected.hist(bins=20, figsize=(12, 10), edgecolor='black')
plt.suptitle("Feature Distributions")
plt.show()
```

```python
# 3. Boxplots to detect outliers
plt.figure(figsize=(12, 6))
sns.boxplot(data=df_selected)
plt.xticks(rotation=90)
plt.title("Boxplot of Features")
plt.show()
```
